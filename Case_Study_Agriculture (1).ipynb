```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data=pd.read_csv("/content/IAS.csv")

# state wise production
state_production=data.groupby('State_Name')['Production'].sum().reset_index()
state_production
```

```python
# in whaich year the production was maximum
data[data['Production']==data['Production'].max()]

```

```python
# seasonwise production
season_production=data.groupby('Season')['Production'].sum().reset_index()
season_production
```

```python
# season wise production including year
season_production=data.groupby(['Season', 'Crop_Year'])['Production'].sum().reset_index()
season_production
```

```python
# suppose i am a farmer which crop should be suitable for me in which season and in which state and district
crop_state=data.groupby(['Crop', 'State_Name', 'District_Name'])['Production'].sum().reset_index()
crop_state

```

```python
# find relation between area and production in line
plt.plot(data['Area'], data['Production'])
plt.xlabel('Area')
plt.ylabel('Production')
plt.title('Area vs Production')
plt.show()

```
